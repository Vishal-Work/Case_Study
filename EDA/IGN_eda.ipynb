# task 4

```python
avg = df["score"].mean()
avg
```

```python
y["score"].mean() #so we can't use y dataframe
```

```python
above_avg=[]
for m in main_platform:
        b=(df["platform"] == m) & (df["score"] > avg)
        n=df[b]
        c = n["score_phrase"].count()
        above_avg.append(c)
print(above_avg)        
```

```python
below_avg=[]
for m in main_platform:
        b=(df["platform"] == m) & (df["score"] <= avg)
        n=df[b]
        c = n["score_phrase"].count()
        below_avg.append(c)
print(below_avg) 
```

```python
total_games=[]
for m in main_platform:
        b=(df["platform"] == m)
        n=df[b]
        c = n["score_phrase"].count()
        total_games.append(c)
print(total_games)
```

```python
t_games=0
for ele in range(0, len(total_games)):
    t_games = t_games + total_games[ele]
print(t_games) 
```

```python
above_avg_pcent=[]

for x in above_avg:
    p=(x/t_games)*100
    above_avg_pcent.append("%.2f" %p)
print(above_avg_pcent)    
```

```python
main_platform
```

```python
plt.figure(figsize=(18,5))

plt.subplot(131)
plt.pie(above_avg_pcent,labels=main_platform,autopct='%0.1f%%')

plt.subplot(132)
plt.plot(main_platform,above_avg_pcent

plt.subplot(133)
plt.bar(main_platform,above_avg_pcent)

plt.show()
```

```python
below_avg_pcent=[]

for x in below_avg:
    p=(x/t_games)*100
    below_avg_pcent.append("%.2f" %p)
print(below_avg_pcent)  
```

```python
plt.figure(figsize=(18,5))

plt.subplot(131)
plt.pie(below_avg_pcent,labels=main_platform,autopct='%0.1f%%')

plt.subplot(132)
plt.plot(main_platform,below_avg_pcent)

plt.subplot(133)
plt.bar(main_platform,below_avg_pcent)

plt.show()
```
